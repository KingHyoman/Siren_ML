### csv 만들기

```python
import csv
```

# training case

```python
f = open('training_Dataset.csv', 'w', newline='')

wr = csv.writer(f)

wr.writerow(['slice_file_name', 'fold', 'classID', 'class'])

start = 1
cnt = 4
cnt_32 = 0
for i in range(1, 4):
  for j in range(1, 24):
    for l in range(-20, 21, 10):
      for k in range(25, 101, 25):
        if cnt_32 == 16:
          cnt_32 = 0
          for num in range(start, start + cnt):
            for volume in range(25, 101, 25):     
              wr.writerow([f'noise_ch{num}_{volume}%.wav', '1', '1', 'none']) 
          start += cnt
        wr.writerow([f'mixed{i}_ch{j}_{k}%_{l}ratio.wav',f'{5}',f'{0}',"alarm"])
        cnt_32 += 1
    
for i in range(0, 4):
  for j in range(1, 5):
    if i == 0:
      wr.writerow([f'{i}-{j}.wav', '2', '1', 'none'])
    else:
      wr.writerow([f'{i}-{j}.wav', '2', '0', 'alarm'])

f.close()
```

#test case

```python
f = open('test_Dataset.csv', 'w', newline='')

wr = csv.writer(f)

wr.writerow(['slice_file_name', 'fold', 'classID', 'class'])

for i in range(1, 4):
  for j in range(1, 33):
    if j%8 == 0:
      for k in range(25, 101, 25):
        for l in range(-20, 21, 10):
          wr.writerow([f'mixed{i}_ch{j}_{k}%_{l}ratio.wav',f'{5}',f'{0}',"alarm"])

for num in range(331, 391):
    for volume in range(25, 101, 25):
        wr.writerow([f'noise_ch{num}_{volume}%.wav', '1', '1', 'none']) 

for num in range(1, 7):
  wr.writerow([f'mixed{num}.wav', '3', '0', 'alarm'])

for num in range(1, 4):
  wr.writerow([f'noise{num}.wav', '3', '1', 'none']) 


f.close()
```
